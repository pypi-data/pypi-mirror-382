```python
import pandas as pd
import pytae as pt
import seaborn as sns
penguins = sns.load_dataset('penguins')
```

```python
df = (
    penguins
    .pipe(pt.group_n,dropna=False)
)
#notice that penguins dataset has multiple rows for same group
df.head()
```

```python
df.long(col='features')
```

```python
df.long(col='features',value='measure')
```

```python
df.long(col='features').wide(col='features')
```

```python
penguins.long(col='features').wide(col='features',aggfunc='max')
```

```python
df=pd.DataFrame({'id':['a','b','c','d','e','','f'],
                  'balance':[10,20,0,21,15,10,25],
                  'country':['sg','cn','ca','np','in','in','in']})

result=df.wide(col='country',value='balance')

result
```

```python
expected_df=df.pivot(index='id', 
                           columns='country', 
                           values='balance'
                        ).reset_index()
expected_df.columns.name = None
expected_df
#notice columns are unique
```
