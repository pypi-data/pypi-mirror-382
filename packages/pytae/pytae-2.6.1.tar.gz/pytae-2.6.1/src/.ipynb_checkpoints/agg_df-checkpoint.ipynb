```python
import numpy as np
import pandas as pd
import pytae as pt
import seaborn as sns
penguins = sns.load_dataset('penguins')
```

```python
df=pd.DataFrame({'id':['a','b','c','d','e','','f','f'],
                  'balance':[10,20,0,21,15,10,20,25],
                  'country':['sg','cn','ca','np','in','in','in','in']})
```

```python
df['id'] = df['id'].replace('', np.nan)
df.agg_df(aggfunc=['sum','min','mean','min','max','n'])
```

```python
df[['id','balance']].agg_df(aggfunc=['sum','n'])
```

```python
penguins.agg_df(aggfunc=['max','sum','n'],dropna=False)
#notice n comes first.
```

```python
penguins.group_x(group=['species',	'island',	'sex'],dropna=True)
```
