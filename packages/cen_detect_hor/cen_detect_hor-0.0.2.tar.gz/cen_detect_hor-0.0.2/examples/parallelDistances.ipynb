```python
import multiprocessing
from Bio import SeqIO
from cen_detect_hor.parallel_distance import build_seqs_distance_matrix_by_chunks, FileSystemChunkStore
from cen_detect_hor.featureUtils import BED_file_to_features, feature_to_seq
from cen_detect_hor.matrix_utils import save_matrix_as_triu, load_matrix_from_triu
```

```python
references = {seq.id : seq for seq in SeqIO.parse("./data/monomers/chr2_centromere_plus2Mb.fasta", "fasta")}
```

```python
monomers_as_seqs = [feature_to_seq(feature, references) for feature in monomers_as_features]
```

```python
#monomer_dists_seq = build_seqs_distance_matrix(monomers_as_seqs)
```

```python
monomer_dists_par = build_seqs_distance_matrix_by_chunks(monomers_as_seqs, num_chunks=32, chunk_store=FileSystemChunkStore("data/submatrices_32/piece_{row}_{col}"))
```
