```python
from collections import defaultdict

def group_emails_by_author(pairs: list[tuple[str, str]]) -> dict[str, list[str]]:
    """
    Recibe una lista de tuplas (author_name, author_email) y devuelve
    un diccionario {nombre: [emails]}.
    Si un email aparece asociado a varios nombres, se conserva
    el nombre más largo como clave final.

    Además, agrupa por el prefijo del email (parte local antes de '@'),
    de modo que 'aaaaa@fhi.co' y 'aaaaa@hisfundation.co'
    se consideren la misma persona y se unifiquen bajo el nombre más largo.
    """
    # Map de parte local (antes de '@') -> mejor nombre (más largo)
    localpart_to_best_name = {}
    # Map de parte local -> conjunto de emails asociados (dominios distintos)
    localpart_to_emails = defaultdict(set)

    for name, email in pairs:
        if not isinstance(name, str) or not isinstance(email, str):
            continue
        name = name.strip()
        email = email.strip()
        if "@" not in email:
            continue
        local = email.split("@", 1)[0].lower()

        current_best = localpart_to_best_name.get(local)
        best_name = name if current_best is None or len(name) > len(current_best) else current_best
        localpart_to_best_name[local] = best_name

        localpart_to_emails[local].add(email)

    # Construimos nombre -> emails, fusionando todas las partes locales
    # que terminen con el mismo mejor nombre
    name_to_emails = defaultdict(set)
    for local, emails in localpart_to_emails.items():
        best_name = localpart_to_best_name[local]
        name_to_emails[best_name].update(emails)

    # Convertimos sets a listas ordenadas
    return {name: sorted(emails) for name, emails in name_to_emails.items()}


# Ejemplo
authors = [
    row
   for row in df[["target_author_name", "target_author_email"]].itertuples(index=False)
] + [
    row
   for row in df[["other_author_name", "other_author_email"]].itertuples(index=False)
]

raw_mailmap = group_emails_by_author(authors)
# {'Bobby': ['bob@example.com'],
#  'Alice': ['alice.work@example.com', 'alice@example.com']}
```
